**4. WheatSeeds Dataset**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Abalone dataset
df = pd.read_csv('/content/wheat_seeds.csv')
df
```

```python
# Data Analysis & Summarization
print("Summary Statistics for Wheat Seeds Dataset:")
print(df.describe())
```

```python
# Statistical Analysis
sns.histplot(df["Perimeter"])
plt.title('Distribution of Perimeter')
plt.xlabel('Perimeter')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='Perimeter', y='Area', data=df)
plt.title('Boxplot of Area by Perimeter')
plt.xlabel('Perimeter')
plt.ylabel('Area')
plt.show()
```

```python
# Correlation Coefficients
correlation_matrix = df.select_dtypes(include='number').corr()
print("Correlation Matrix for Wheat Seeds Dataset:")
print(correlation_matrix)
```

```python
# NULL Values
print("NULL Values in Each Column:")
print(df.isnull().sum())
df['Area'].fillna(df['Perimeter'].mean(), inplace=True)
print("NULL Values After Imputation:")
print(df.isnull().sum())
```
