**5. Boston HousePrice Dataset**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Abalone dataset
df = pd.read_csv('/content/boston.csv')
df
```

```python
# Data Analysis & Summarization
print("Summary Statistics for Boston Housing Dataset:")
print(df.describe())
```

```python
# Statistical Analysis
sns.histplot(df["INDUS"])
plt.title('Distribution of Median House Price (INDUS)')
plt.show()
```

```python
sns.boxplot(x='INDUS', y='TAX', data=df)
plt.xlabel('INDUS')
plt.ylabel('TAX')
plt.title('Boxplot of TAX by INDUS')
```

```python
# Correlation Coefficients
correlation_matrix = df.corr()
print("Correlation Matrix for Boston Housing Dataset:")
print(correlation_matrix)
```

```python
# NULL Values
print("NULL Values in Each Column:")
print(df.isnull().sum())
df.fillna(df.mean(), inplace=True)
print("NULL Values After Imputation:")
print(df.isnull().sum())
```
